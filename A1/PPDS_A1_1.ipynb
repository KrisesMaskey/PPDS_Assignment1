```python
articles = data['articles']

query_template = f"""
                    INSERT IGNORE INTO 
                    {db_name}.{table_name}(news_title, news_author, news_url, news_content) 
                    VALUES (%s, %s, %s, %s)
                  """

for article in articles:
    news_title = article['title']
    news_author = article['author']
    news_url = article['url']
    news_content = article['content']

    print("Inserting article", news_title, "with", news_url, "as URL")
    query_parameters = (news_title, news_author, news_url, news_content)
    engine.execute(query_template, query_parameters)
```
