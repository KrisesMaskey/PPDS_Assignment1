```python
#Store Sentiment Scores in Database
new_table_name = f"{suffix}_sentiment_db"

# Create a table
create_table_query = f"""CREATE TABLE IF NOT EXISTS {db_name}.{new_table_name} 
                                 (news_url varchar(250),
                                 news_sscore double(64, 30),
                                 PRIMARY KEY(news_url)
                                 )"""
engine.execute(create_table_query)

query_template = f"""
                    INSERT IGNORE INTO 
                    {db_name}.{new_table_name}(news_url, news_sscore) 
                    VALUES (%s, %s)
                  """
for data in data_entries:
    news_url = data['retrieved_url']
    news_sscore = float(data['sentiment']['document']['score'])

    print("Inserting Sentiment Score to", news_url, "=", news_sscore)
    query_parameters = (news_url, news_sscore)
    engine.execute(query_template, query_parameters)
```
